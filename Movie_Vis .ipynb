## Scatter Plot

```python
plt.scatter(movies_df['budget'], movies_df['revenue'])
plt.title('Budget vs. Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

## Histogram

```python
plt.hist(movies_df['vote_average'], bins=30)
plt.title('Distribution of Movie Score')
plt.xlabel('Vote')
plt.ylabel('Count')
plt.show()
```

## Line Plot

```python
movies_df['date_x'] = pd.to_datetime(movies_df['release_date'])

Revenue_over_time = movies_df.groupby(movies_df['date_x'].dt.year)['revenue'].mean()

sns.lineplot(x=Revenue_over_time.index, y=Revenue_over_time.values)
plt.title('Revenue of Movies Over Time')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.show()
```

## Bar Plot

```python
credits_df = credits_df.rename(columns={'movie_id': 'id'})
data = pd.merge(movies_df, credits_df, on='id')
sns.set_theme(style="whitegrid")
plt.figure(figsize=(16, 15))
sns.barplot(x="revenue", y="original_language", data=data, label="Revenue", color="b", ci=None)
sns.barplot(x="budget", y="original_language", data=data, label="Budget", color="g", ci=None)
plt.legend(loc="lower right", frameon=True)
plt.xlim(0, 10000 * 15000)
plt.xlabel("Language Effect on Sales")
plt.ylabel("")
sns.despine(left=True, bottom=True)
plt.show()
```

## Scatter Plots

```python
top_movies = movies_df.nlargest(20, 'revenue')

plt.figure(figsize=(12, 8))

scatter = sns.scatterplot( data=top_movies, x='vote_average', y='revenue', color='blue', alpha=0.7 )
for i in range(len(top_movies)):
    plt.text( x=top_movies['vote_average'].iloc[i], y=top_movies['revenue'].iloc[i] + 1e7, s=top_movies['title'].iloc[i], 
             fontsize=8, ha='center', alpha=0.9 )

plt.title('Top 20 Movies: Relationship Between Rating and Revenue', fontsize=16)
plt.xlabel('Rating (Average)', fontsize=12)
plt.ylabel('Revenue (Millions)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
movies_df['main_genre'] = movies_df['genres'].str.split(',').str[0]
plt.figure(figsize=(10, 6))
sns.scatterplot( data=movies_df, x='vote_average', y='revenue', hue='main_genre', palette='muted', alpha=0.7 )
plt.title('Rating vs Revenue by Main Genre', fontsize=14)
plt.xlabel('Rating (Average)', fontsize=12)
plt.ylabel('Revenue (Millions)', fontsize=12)
plt.legend(title='Main Genre', bbox_to_anchor=(1, 1), fontsize=10)
plt.tight_layout()
plt.show()
```
